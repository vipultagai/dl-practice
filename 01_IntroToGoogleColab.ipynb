```
file.close()
```

```
with open('imports-85.data', 'r') as file:
    print(file.readline())
```

```
with open('imports-85.data', 'r') as file:
    print(file.read())
```

```
with open('imports-85.data', 'r') as file:
    for line in file:
        print(line)
```

```
with open('imports-85.data', 'r') as file:
    for line in file:
        for word in line.split(','):
            print(word)
        print('------')
```
